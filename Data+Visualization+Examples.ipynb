Exercise 3. A. Create a 3d scatter plot where the $x$-values range from $-2$ to $4$, the $y$-values are the cosine-squared of $x$, and the $z$ values are the exponential of the sine of the difference of $x$ and $y$.  Set the step size for $x$ to be at most $0.01$.  
   B. Using the `meshgrid` method, graph a wireframe *hyperbolic paraboloid* for $a=\sqrt{3}$, $b=2$.  (Have the $xy$ values range from at least $-10$ to $10$ to see the structure properly.)

```python
# 3A
x = np.arange(-2,4,0.01)
y = np.cos(x)**2
z = np.exp(sin(x-y))

fig3a = m3d.Axes3D(plt.figure())
fig3a.scatter3D(x,y,z, color='black')
plt.show()
```

```python
# 3B
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
xx, yy = np.meshgrid(x, y)
Z = xx + 1.5*yy + 2
figplane = m3d.Axes3D(plt.figure())
figplane.plot_wireframe(xx, yy, Z, rstride=10, cstride=10);
plt.show();
```
